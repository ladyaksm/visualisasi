### Cleaning Data

## Exploratory Data Analysis (EDA)

```python
# EDA untuk Pertanyaan 1: Kategori Produk Paling Banyak Dibeli

# Pola pembelian kategori produk dari waktu ke waktu
order_items_with_date = (order_items_df
                         .merge(product_df, on="product_id", how="left")
                         .merge(orders_df, on="order_id", how="left"))

# Convert 'order_purchase_timestamp' to datetime
order_items_with_date["order_purchase_timestamp"] = pd.to_datetime(order_items_with_date["order_purchase_timestamp"])

category_trend = (order_items_with_date
                  .groupby([order_items_with_date["order_purchase_timestamp"].dt.to_period("M"), "product_category_name"])
                  .size()
                  .reset_index(name="order_count"))

print("\nTren pembelian kategori produk per bulan:")
print(category_trend.head(10))


# Kategori produk dengan jumlah pembelian terbanyak
top_categories = (order_items_df
                  .merge(product_df, on="product_id", how="left")
                  .groupby("product_category_name")
                  .size()
                  .reset_index(name="total_orders")
                  .sort_values(by="total_orders", ascending=False))

print("Top 10 kategori produk paling banyak dibeli:")
print(top_categories.head(10))

# Rata-rata harga produk per kategori
avg_price_per_category = (order_items_with_date
                          .groupby("product_category_name")["price"]
                          .mean()
                          .reset_index()
                          .sort_values(by="price", ascending=False))

print("\nRata-rata harga produk per kategori:")
print(avg_price_per_category.head(10))
```

```python
# EDA untuk Pertanyaan 2: Pola Geografis dalam Jumlah Pesanan


# Kota dengan jumlah pesanan terbanyak
orders_with_city = orders_df.merge(customers_df, on="customer_id", how="left")

top_cities = (orders_with_city
              .groupby("customer_city")
              .size()
              .reset_index(name="total_orders")
              .sort_values(by="total_orders", ascending=False))

print("\nTop 10 kota dengan jumlah pesanan terbanyak:")
print(top_cities.head(10))

# Hubungan antara lokasi seller dan pelanggan
order_seller_location = (order_items_df
                         .merge(orders_df, on="order_id", how="left")
                         .merge(customers_df, on="customer_id", how="left")
                         .merge(sellers_df, on="seller_id", how="left"))

# Jumlah pesanan per hubungan seller dan pelanggan berdasarkan kota
seller_customer_city = (order_seller_location
                        .groupby(["seller_city", "customer_city"])
                        .size()
                        .reset_index(name="order_count"))

print("\nContoh hubungan seller dan customer berdasarkan kota:")
print(seller_customer_city.head(10))

# Distribusi pesanan berdasarkan provinsi pelanggan
orders_by_state = (orders_with_city
                   .groupby("customer_state")
                   .size()
                   .reset_index(name="total_orders")
                   .sort_values(by="total_orders", ascending=False))

print("\nDistribusi pesanan berdasarkan provinsi pelanggan:")
print(orders_by_state)
```

**Insight:**
Pertanyaan 1:

- Menghitung kategori produk yang paling sering dibeli (top_categories)
- Melihat tren pembelian kategori produk per bulan (category_trend)
- Menampilkan rata-rata harga produk dari setiap kategori (avg_price_per_category)

Pertanyaan 2:

- Menghitung kota dengan jumlah pesanan terbanyak (top_cities)

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
#  Visualisasi Tren Pembelian Kategori Produk Per Bulan
# Konversi order_purchase_timestamp ke datetime jika belum
orders_df["order_purchase_timestamp"] = pd.to_datetime(orders_df["order_purchase_timestamp"])

# Gabungkan dengan order_items, product, DAN product_category_df untuk mendapatkan kategori
merged_data = order_items_df.merge(orders_df, on="order_id").merge(product_df, on="product_id").merge(product_category_df, on="product_category_name", how="left")

# Ambil hanya kolom yang penting
merged_data["purchase_month"] = merged_data["order_purchase_timestamp"].dt.to_period("M")

# Hitung jumlah transaksi per kategori per bulan
category_trends = merged_data.groupby(["purchase_month", "product_category_name_english"])["order_id"].count().reset_index()

# Visualisasi tren pembelian kategori produk tertentu
plt.figure(figsize=(14, 7))

# top_categories didefinisikan dan berisi kolom 'product_category_name_english'

top_categories = category_trends.groupby("product_category_name_english")["order_id"].sum().reset_index().sort_values(by="order_id", ascending=False)

top_categories_list = top_categories["product_category_name_english"].head(5).tolist()  # Ambil 5 kategori teratas

for category in top_categories_list:
    data = category_trends[category_trends["product_category_name_english"] == category]
    plt.plot(data["purchase_month"].astype(str), data["order_id"], marker="o", label=category)

plt.xlabel("Bulan")
plt.ylabel("Jumlah Transaksi")
plt.title("Tren Pembelian Produk per Kategori dalam Beberapa Bulan Terakhir")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Visualisasi Kategori Produk Paling Banyak Dibeli
plt.figure(figsize=(12, 6))

# Use the original top_categories DataFrame or recreate it:
top_categories = (order_items_df
                  .merge(product_df, on="product_id", how="left")
                  .groupby("product_category_name")
                  .size()
                  .reset_index(name="total_orders")
                  .sort_values(by="total_orders", ascending=False))

# Now you can use top_categories for the barplot:
sns.barplot(data=top_categories.head(10), x="total_orders", y="product_category_name", palette="coolwarm")
plt.xlabel("Total Orders")
plt.ylabel("Product Category")
plt.title("Top 10 Kategori Produk Paling Banyak Dibeli")
plt.show()
```

```python
# Visualisasi Rata-rata Harga Produk Per Kategori
plt.figure(figsize=(12, 6))
sns.barplot(data=avg_price_per_category.head(10), x="price", y="product_category_name", palette="magma")
plt.xlabel("Rata-rata Harga (IDR)")
plt.ylabel("Kategori Produk")
plt.title("Top 10 Kategori dengan Rata-rata Harga Tertinggi")
plt.show()
```

### Pertanyaan 2:

```python
# Visualisasi Hubungan Seller dan Pelanggan Berdasarkan Kota (Heatmap)
import matplotlib.colors as mcolors # Import the necessary module

pivot_table = seller_customer_city.pivot_table(index="seller_city", columns="customer_city", values="order_count", fill_value=0)

# Filter kota dengan jumlah transaksi terbanyak
min_orders = 50  # Batasan minimum transaksi
filtered_pivot = pivot_table.loc[pivot_table.sum(axis=1) > min_orders, pivot_table.sum(axis=0) > min_orders]

#  Jika masih terlalu besar, ambil hanya top 20 seller & customer cities
top_seller_cities = seller_customer_city["seller_city"].value_counts().head(20).index
top_customer_cities = seller_customer_city["customer_city"].value_counts().head(20).index
filtered_pivot = filtered_pivot.loc[filtered_pivot.index.intersection(top_seller_cities),
                                    filtered_pivot.columns.intersection(top_customer_cities)]

# Plot heatmap dengan skala log
plt.figure(figsize=(12, 8))
sns.heatmap(filtered_pivot, cmap="coolwarm", linewidths=0.5, norm=mcolors.LogNorm()) # Use the imported class

# Tambahkan judul & label
plt.title("Hubungan Seller dan Pelanggan Berdasarkan Kota")
plt.xlabel("Customer City")
plt.ylabel("Seller City")
plt.xticks(rotation=90)
plt.yticks(rotation=0)

plt.show()
```

```python
# Visualisasi Kota dengan Pesanan Terbanyak
plt.figure(figsize=(12, 6))
sns.barplot(data=top_cities.head(10), x="total_orders", y="customer_city", palette="viridis")
plt.xlabel("Total Orders")
plt.ylabel("Customer City")
plt.title("Top 10 Kota dengan Jumlah Pesanan Terbanyak")
plt.show()
```

```python
# Visualisasi Distribusi Pesanan Berdasarkan Provinsi
plt.figure(figsize=(12, 6))
sns.barplot(data=orders_by_state, x="customer_state", y="total_orders", palette="coolwarm")
plt.xlabel("State")
plt.ylabel("Total Orders")
plt.title("Distribusi Pesanan Berdasarkan Provinsi")
plt.xticks(rotation=45)
plt.show()
```
